```python
data_cleaned = data.drop (['Data', 'Hora UTC', 'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)', 'PRESSAO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)', 'PRESSAO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)', 'RADIACAO GLOBAL (Kj/m²)', 'TEMPERATURA DO PONTO DE ORVALHO (°C)', 'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)', 'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)','VENTO, DIREÇAO HORARIA (gr) (° (gr))', 'VENTO, RAJADA MAXIMA (m/s)', 'VENTO, VELOCIDADE HORARIA (m/s)'], axis=1)
data_cleaned.head()
```

```python
data_cleaned.isnull().sum()
```

```python
def num_missing(x):
  return sum(x.isnull())
```

```python
import numpy as np

data_cleaned.apply(num_missing, axis=0) #axis=0 defines that function is to be applied on each column
```

```python
# Calculo da Mediana
median_prec = data['PRECIPITAÇAO TOTAL, HORARIO (mm)'].median()  
# Substitui valores nulos pela mediana
data_cleaned['PRECIPITAÇAO TOTAL, HORARIO (mm)'].fillna(median_prec, inplace=True)


median_tempb = data['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].median()  
# Substitui valores nulos pela mediana
data_cleaned['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].fillna(median_prec, inplace=True)


median_temmax = data['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (°C)'].median()  
# Substitui valores nulos pela mediana
data_cleaned['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (°C)'].fillna(median_prec, inplace=True)

median_tempmin = data['TEMPERATURA MINIMA NA HORA ANT. (AUT) (°C)'].median() 
# Substitui valores nulos pela mediana
data_cleaned['TEMPERATURA MINIMA NA HORA ANT. (AUT) (°C)'].fillna(median_prec, inplace=True)


median_ummax = data['UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'].median()
# Substitui valores nulos pela mediana
data_cleaned['UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'].fillna(median_prec, inplace=True)

median_ummin = data['UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)'].median()
# Substitui valores nulos pela mediana
data_cleaned['UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)'].fillna(median_prec, inplace=True)


median_umrel = data['UMIDADE RELATIVA DO AR, HORARIA (%)'].median()
# Substitui valores nulos pela mediana
data_cleaned['UMIDADE RELATIVA DO AR, HORARIA (%)'].fillna(median_prec, inplace=True)

data_cleaned.apply(num_missing, axis=0) #axis=0 defines that function is to be applied on each column
```
